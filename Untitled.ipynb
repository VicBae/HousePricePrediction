Data Cleaning

```python
# Check for row duplication
X.duplicated().sum()  # No duplication

# Check for NaN values
NaNlist = X.isna().sum()

# Listing features with NaN values only
NaNlist = NaNlist[NaNlist.values!=0]
# Sort it in descending order and get the percentage that NaN value takes 
NaNlist_percentage = NaNlist.sort_values(ascending=False) / X.shape[0]
NaNlist_percentage
```

```python
#Visualize it 
plt.figure(figsize=(10,5))
sns.barplot(x = NaNlist_percentage.index, y = NaNlist_percentage.values * 100);
plt.title("Percentage of NaN values");
plt.ylabel("%");
plt.xticks(rotation=90);
```

I decided to drop any features that have more than 10% of NaN values because it is hard to fill NaN values and we will have very weak basis to do so.

```python
dropfeatures = NaNlist[NaNlist_percentage>0.1].index.values
X = X.drop(dropfeatures,axis=1) # We dropped 6 columns that had more than 10% of NaN values
```

Categorical Features

```python
categorical_features = X.dtypes[X.dtypes=="object"].index.values
categorical_features
```

Initially I wanted to apply VarianceThreshold to see which columns are useless for prediction however it doesn't support categorical values and if it has NaN values it won't work.

We will do this manually.

```python
X.MSZoning.value_counts()[0]
```

```python
# some features will not be helpful for prediction if there is very small variance in them
frequency=[]
for i in categorical_features:
    frq = X[i].value_counts()[0] / X.shape[0]
    frequency.append([i,frq])
    
frequency=np.array(frequency)    
vis_frequency = pd.DataFrame(index=frequency[:,0],data=frequency[:,1],columns=['Frequency'])
vis_frequency=vis_frequency.Frequency.sort_values(ascending=False)


plt.figure(figsize=(20,5))
sns.barplot(x = vis_frequency.index, y = vis_frequency.values.astype(float));
plt.title("Percentage of NaN values");
plt.ylabel("%");
plt.xticks(rotation=90);
```

```python
type(vis_frequency)
```

```python
plt.figure()
plt.bar(vis_frequency.index, vis_frequency.Frequency)
plt.show()
```

Numerical Features

```python
numeric_features = X.dtypes[X.dtypes!="object"].index.values
numeric_features
```

```python
from sklearn.feature_selection import VarianceThreshold


sel = VarianceThreshold(threshold=0)
Xfixed = sel.fit_transform(X[numeric_features]) # won't work until we fix NaN
```

```python
from sklearn.metrics import adjusted_mutual_info_score

mutual_info_scores = []

for feature in numeric_features:
    # Calculate the mutual info score
    m_info = adjusted_mutual_info_score(X[feature], y)
    # Append it to our list
    mutual_info_scores.append(m_info)

plt.figure()
plt.bar(numeric_features, mutual_info_scores)
plt.ylabel('Mutual info shared with Target')
plt.xticks(rotation=90)
plt.plot()
```
